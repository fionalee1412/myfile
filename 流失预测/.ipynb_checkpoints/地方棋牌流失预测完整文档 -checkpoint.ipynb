## 项目背景
> 目前运营做的流失召回，大部分是针对已经流失的用户，由于用户大部分已经流失了，再次挽回的效果并不好，就像分手的情侣一样，已经分手大半个月了，你再去挽回一样。  
更好的做法是，在用户已经有流失的迹象时，马上采取相应的运营手段，这样才能更有效地防止流失。  
我们这次的目的主要是找出未来7天内不登录的用户，这些用户不一定流失，但是不登录玩牌的话，对游戏的贡献也相当于0，所以有必要采取运营措施

## 样本不平衡解决方案 
#### 通用的解决方案有：  
* 不均衡随机采样：包括对流失过采样，或者对未流失欠采样，使得两者比例平衡
    * 一般的教程是假设数据集是固定的，欠采样和过采样都是在同一个数据集下，通过重复采样等方式达到均衡的效果
    * 这里我们有一点很好的地方，我们可以从数据库中多抽取出流失用户，使得流失和未流失的比例平衡，从而避免了欠采样的缺陷
* 如果数据集是固定的（比如参加kaggle的竞赛），更好的办法是采用informed欠采样，主要的算法有以下两个：
    * EasyEnsemble算法：思想是训练多个基分类器（从多数样本中独立随机抽样，结合少数的那个联合训练，再讲这些基分类器组合）
    * BalanceCascade算法：使用之前已形成的集成分类器来为下一次训练选择多数样本，然后再进行欠采样
* 还可以采用最近邻的思想来对多数样本欠采样
    * 这种方法的思想是基于数据分布，对多数样本的数据，找出其最近邻的样本，如果这个样本最近邻中有两个以上是少数类样本，那么就删掉这个多数类样本点
* 还可以自己合成数据，这方面有名的技术是SMOTE过抽样技术
    * 区别于简单的复制少数样本的过采样，这个技术利用了少数样本之间的相似性来构建数据，类似于直接复制生产的是克隆人，那这种方法就是在复制的基础上加上了变异
    
#### 总结：  
* 一般机器学习的参考书里面提供的样本都是比较平衡的，所以基本上也没有提到过样本不平衡要怎么办；但是，在真实的数据里面，不平衡的样本才是常态，比如流失和异常检测等。
* 网上很多讲解样本不平衡的解决方案也是很简单的告诉你，要过采样或者欠采样，或者自己合成数据；这些人估计都没有做过真实的案例（除了大部分的都是直接复制别人的文章之外，还有的是只参加过kaggle竞赛，自己也没有碰过真实数据，因为根本就未提及到数据量够的话是否还需要这样做）
* 在实际情况里面，特别是积累了大量数据的公司，比如我们要识别作弊用户，虽然在整体的样本分布里面，作弊用户的比例是很低的，但是我们可以直接从hive里面提取出这些用户，再随机提取正常用户，使得样本是平衡的
* 我最开始做的时候，也没有考虑到样本的问题，直接用了准确度来计算，导致准确度异常高，因为未流失的占大部分，即便我们一个流失用户都没有预测到，那么准确度也有80%
* 对于机器学习，很多人都是通过参加kaggle竞赛来训练自己，但是这样存在很大的问题：
    1. 无法训练特征采集，然而项目做多了会发现特征才是最重要的
    2. 模型大部分也是调用包（所以很多机器学习人员也称为调包侠），整个过程中，唯一做的事可能就是调调算法的参数
    3. kaggle比赛的优胜者做得好的地方也在于特征的构建（因为好的算法基本也就那几种）
    4. 很多人只是学到了整个机器学习的过程，加上各种算法的原理（当然这个也很重要），就以为自己已经掌握了机器学习
* 一般研究生学习的路径是先学理论，再通过参加竞赛来提高算法能力。对于我们这种上班族来说，有一个很好的优势是，我们有大量真实的数据，可以构建出各种解决方案来解决实际问题；
* 所以我们可以先跑流程，然后在实际过程中再去补算法知识，虽然刚开始的时候学起来会比较累，但由于有了实际应用场景，在学算法的时候，就能够结合案例来学，理解起来也更快，而且可以边学边用，并且不断调整，掌握得也更牢固
* 所以虽然有点后悔大学没有好好学习高数和线性代数，但是想想即便当时学好了，估计现在也早忘了
* 机器学习的发展也是相当快，现在神经网络、深度学习都是用得比较普遍的了，最近又出了强化学习，所以学习能力是很重要的，虽然和机器学习一样，不需要我们去想算法（这是真正数学大牛们的事），但是至少要知道算法原理，以及怎么用

#### 递归特征消除
递归特征消除的主要思想是反复的构建模型（如SVM或者回归模型）然后选出最好的（或者最差的）的特征（可以根据系数来选），把选出来的特征放到一遍，然后在剩余的特征上重复这个过程，直到所有特征都遍历了。这个过程中特征被消除的次序就是特征的排序。因此，这是一种寻找最优特征子集的贪心算法。

```python
# Fit model using each importance as a threshold
from numpy import sort
from sklearn.metrics import f1_score
thresholds = sort(XGB.feature_importances_)
thresh_list = {}
f1 = f1_score(y_test, predictions,average='binary')
thresh_list_f1 = {}
for thresh in thresholds:
    # select features using threshold
    selection = SelectFromModel(XGB, threshold=thresh, prefit=True)
    select_X_train = selection.transform(X_train)
    # train model
    selection_model = XGBClassifier()
    selection_model.fit(select_X_train, y_train)
    # eval model
    select_X_test = selection.transform(X_test)
    y_pred = selection_model.predict(select_X_test)
    predictions = [round(value) for value in y_pred]
    accuracy = accuracy_score(y_test, predictions)
    thresh_list[select_X_train.shape[1]] = accuracy*100.0+thresh
    f1 = f1_score(y_test, predictions,average='binary')
    thresh_list_f1[select_X_train.shape[1]] = f1*100.0+thresh
```

## 模型分值

```python

def algorithm(X,y,score):
    num_folds = 10
    scoring = score
    seed = 7
    kfold = KFold(n_splits=num_folds,random_state=seed)
    results = []

    ## 普通算法
    models = {}
    # 逻辑回归
    models['LR'] = LogisticRegression()
    # 线性判别分析
    models['LDA'] = LinearDiscriminantAnalysis()
    # K近邻
    models['KNN'] = KNeighborsClassifier()
    # 分类与回归树
    models['CART'] = DecisionTreeClassifier()
    # 贝叶斯分类器
    models['NB'] = GaussianNB()
    # 支持向量机
    models['SVM'] = SVC()
    
    for key in models:
        cv_results = cross_val_score(models[key],X,y,cv=kfold,scoring=scoring)
        results.append(cv_results)
        print('%s: %f (%f)'% (key,cv_results.mean(),cv_results.std()))
        
    ## 集成算法
    ensembles = {}
    ensembles['ScaledAB'] = Pipeline([('Scaler',StandardScaler()),('AB',AdaBoostClassifier())])
    ensembles['ScaledGBM'] = Pipeline([('Scaler',StandardScaler()),('GBM',GradientBoostingClassifier())])
    ensembles['ScaledRF'] = Pipeline([('Scaler',StandardScaler()),('RFR',RandomForestClassifier())])
    ensembles['ScaledET'] = Pipeline([('Scaler',StandardScaler()),('ETR',ExtraTreesClassifier())])
  
    for key in ensembles:
        cv_result = cross_val_score(ensembles[key],X,y,cv=kfold,scoring=scoring)
        results.append(cv_result)
        print('%s: %f (%f)'% (key,cv_result.mean(),cv_result.std()))
        
    # 通过箱线图显示数据的分布情况
    fig = plt.figure(figsize=(18, 8), dpi=200)
    ax = fig.add_subplot(111)
    plt.boxplot(results);
    ax.set_xticklabels(list(models.keys())+list(ensembles.keys()));
```
