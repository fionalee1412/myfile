## 目标人群
不同的棋牌游戏类型对应不同类型的玩家  
#### 按照游戏类型，分为：  
* 策略类：麻将类、字牌类、斗地主、德州、大连打滚子、五十K、象棋、军旗、五子棋等
* 金流类：斗牛、焖鸡、十三张、骰子、三公、炸金花、百家乐等

#### 按照实体类型，分为：
* 扑克牌类：斗地主、德州、大连打滚子、斗牛等
* 麻将类：麻将
* 棋类：象棋、五子棋、军旗等
* 其他：骰子、字牌等

#### 按照游戏来源，分为：
* 中国：麻将、斗地主、象棋、军旗、斗牛、焖鸡、炸金花等
* 港澳台：德州、梭哈、当地麻将、百家乐等
* 东南亚：三公、十三张、骰子等
* 欧美：德州、二十一点、国际象棋等

#### 即便是同一种类型的游戏，也会有难度和娱乐性上的差异，从而造成用户群体上的差异，例如：  
* 德州从美国开始流行，通过电视节目、名人牌手等，在上层阶级比较流行，加上具有和做投资类似的特点（比如概率和博弈等），所以非常受金融圈人士的欢迎。
* 麻将相传起源于中国明朝，不同的地区会有一些规则上的差异，一般来说，一个地区的人会更加偏向于玩自己当地的麻将，所以具有很强的地域特征。
* 斗地主的历史比麻将要短得多，起源于湖北，自20世纪90年代初在皖南、江苏的部分地区出现斗地主后，斗地主游戏迅速向全国展开，目前已和德州、百家乐成为世界上三种最流行的扑克游戏（百度 百科上说的）。并没有特别多的地区玩法，但是也有很多衍生玩法，比如二人斗地主、四人斗地主、癞子斗地主等。一般的棋牌创业小公司（比较正规的那种），最开始做的一般是斗地主。
* 字牌是一种古老的棋牌游戏，玩法和麻将比较像，但是携带起来比麻将要方便不少；发源于四川乐山五通，在广西、四川和湖南地区很受欢迎，在桂林尤其受欢迎
* 棋类游戏策略性非常强，不像牌类游戏由于不确定性带来的刺激，是纯粹考验技术的游戏，而且多数为二人博弈，规则多数比牌类游戏要复杂
* 金流类游戏，则是几乎完全的概率性游戏，几乎不含有策略，但是规则的差异可以衍生出各种玩法，但是本质都是比大小。稍微掌握一点概率知识的人都不应该参与这种类型的游戏，但是就和博彩一样，用户群相当广。从心理上简单分析，赢了钱会想赢更多钱，输了钱会想着下一把赢回来，于是只要开始游戏，便停不下来，而且由于这种游戏规则都非常简单，每一局时间也特别短，可能不自觉地就会玩很久。虽然不能一概而论地说这些用户的阶层低，但是普遍是自控力较低的。这种游戏也是最能给游戏公司创造收益的，当然也是各国政府严厉打击的对象。
