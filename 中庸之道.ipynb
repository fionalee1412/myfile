中庸之道  
中庸意味着合适的极致  
什么是合适的极致呢？用机器学习类比，就是拟合度非常好，既没有过拟合，也没有欠拟合  
审美也是一样的，人们不喜欢小眼睛，大脸盘，于是使劲把眼睛弄大，削骨隆下巴，造成了另一种丑  
运动也是一样，我之前也觉得多运动总是好的，于是早上锻炼，晚上还要练肌肉，周末去打拳、练瑜伽，结果下来，身体并没有变好，反而更加地虚弱。练健美的人力气虽大，但是不见得能打过肌肉更少的拳击手。记得在看龙珠的时候，当时贝吉塔和特兰克斯进入时间和空间的房间，特兰克斯采取的修行方法是将肌肉最大化，力量值是上升了，但是速度也因此降下来了；悟空采用的是更加平衡的方法，他同时训练了力量和速度。  

财富也是如此，钱不是越多越好，你为了追求金钱，牺牲了时间和身体，最终宇宙的平衡法则会让你挣到的钱补在身体和时间上  

大自然的守恒规则就是中庸之道  

佛教追求的无我，也是中庸之道的体现，无喜无悲，无我相无众生相，都是站在一个中间点上。以前不理解不以物喜，不以己悲，其实也正常，没有到那个境界，其实这句话说的就是无我，就是中庸。  

宇宙的法则用人类的语言是不容易描述的，大家都说佛教不好理解，其实佛教已经很尽力地解释了这个世界，真正地悟道不一定要读佛经，也更不用出家，看懂这个世界就悟道了。  

悟道不是什么事都不做了，而是能在做事的时候感受到自己在做这件事，所谓吃饭是禅，睡觉也是禅，工作是禅，放松也是禅；事物本身没有好坏之分，是人类的分别心让事有了好坏对错，是人类的分别心让世间有了美丑  

现代社会总是追求极致，但是真正领悟到极致的人很少。真正的极致一定是中庸的，所谓的和谐，也是中庸。
