### 项目背景
* 实现自动化运营：公司已有一些仍有活跃和付费的游戏，但是已没有人力去维护运营；我们可以结合机器学习和自动化推送，自动运营这些项目，提高收益的同时又不增加成本
* 实现精准化运营：目前运营的召回是找一批用户去发推送，但是并没有结合每个用户的特征和玩牌习惯，如果能够针对性地发送，应该能够提高点击率
* 结合流失预测：找到预流失用户之后，针对性地发送推送，促进活跃和预防流失

### 项目目的
* 将机器学习预测流失、用户聚类、用户特征分析，结合自动化推送，跑通整套流程，实现自动化运营

### 项目风险
* 用户有可能觉得推送是一种打扰，反而卸载了游戏
    * 控制推送的频率，假如用户对推送并没有反应，应该停止对这个用户的推送
* 能够收到推送的用户量太少，效果不明显
    * 待流程跑通之后，找人多一些的游戏使用
* 公司不同意我们使用自动化运营
    * 以效果来证明
* 苹果推送不会返回用户是否有点击推送（我记得应该也是没有返回是否某个用户有没有推送成功）
    * 这个没办法了，只能统计全部的推送了
* 影响到正常的推送
    * 新建一个通道，或者观察一下是否有并发堵塞的风险

### 项目收益
* 大大减少人工运营成本，可以让运营将更多的精力投入到活动运营上去
* 可实现对用户的精准个性化运营

### 内容
#### 将用户最近玩牌7日的玩牌时段最多的那个时段加入到画像表中
#### 新增促活
* 定义：对当日新增用户，根据其新增当日玩牌情况进行聚类和特征分析
* 流程：
    1. 随机分成两组，一组在新增第二日发送推送，第二组不发送
    2. 观察两组用户的第二日玩牌情况
    
#### 已流失用户召回
* 定义：对于已经流失（超过30日未登录的用户（或者观察一下这个游戏的回流时间来确定））进行推送召回，分析其之前的玩牌习惯，聚类成 不同的用户群，分析其玩牌特征，制定不同的推送召回内容；再根据用户之前最常登录玩牌的时间段，确定每个人的最佳推送时间
* 流程： 
    1. 第一次将所有的流失用户（超过30日未登录）找出来，放在一张表里面
    2. 对这些用户做分类或者聚类，按照玩牌习惯等定制不同的召回文案
    3. 根据用户的玩牌时段，在指定的时间段给用户发推送
    4. 将推送成功的用户放入一张表，不成功的放入一张表
    5. 统计推送成功的用户有多少被召回的

#### 防止用户流失
* 定义：预测当日，前7日有活跃的用户（除去新增用户），在预测到预测日第二天将要流失的，在第二天发送推送；但是如果用户在7日内已经发送过召回，那么即便这次又预测到了这个用户有流失，那也不能发送推送
* 流程：
    1. 建立流失预测模型，将预测出即将流失的用户放入一张表
    2. 针对不同类型的用户，建立几套针对性的召回文案
    3. 将这些用户划分到上面几类用户中去
    4. 根据用户的玩牌时间段，在指定的时间段发送推送
    5. 将发送推送成功的用户放入一张表，不成功的放入一张表


### 效果分析
* 建立A/B测试，将能够收到推送的，和不能收到推送的分为两部分，两者各个层次（生命周期、用户价值分等）的用户数需要基本一致（不采用随机划分的原因在于，担心能收到推送的用户量本来就少，再一分就没几个人了），之后统计这两部分用户的活跃频率、玩牌次数、付费次数等
* 将推送成功的用户在推送之后的行为与之前进行对比
* 根据之前的数据，建立预测模型，将模型运行之后的效果与预测效果进行对比
