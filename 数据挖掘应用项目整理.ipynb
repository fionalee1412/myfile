### 活动运营
> 运营为了促进新增、活跃、付费等做的一系列活动，线上的做法是在游戏内部，用户完成相应的任务，或者是参与抽奖活动，还有的是和其他品牌合作给用户发奖励。通知的手段有：垂直媒体、平台私信、推送、短信、QQ群、电话、邮件等

> 还有一种活动，是为了回收游戏币做的

* 转换为数据分析问题：
    * 找到成本和效果的最优平衡点，以最小的成本得到最好的效果（可根据之前的数据、本次的用户数等特征，预估本次活动的效果，从而找到成本最优解）
    * 活动效果分析（通过AB测试、假设检验等方法，评估本次活动实际的收益）
    

### 精准运营
> 给不同类型的玩牌用户做不同的运营，使运营成本最低，效果最大

* 转换为数据分析问题：
    * 根据用户玩牌习惯等特征，运用聚类的方法，自动将用户分成不同的群体

### 流失召回
> 对不活跃的用户通过推送、短信等方式进行召回

* 转换为数据分析问题：
    * 对可能流失的用户做预测，找出预流失用户

### 防作弊
> 对于不正常用户的识别处理，例如：广告、刷号、通牌作弊等

* 转换为数据分析问题：
    * 对上传的头像做图像识别
    * 对广告、刷号的用户做模式识别
    * 对通牌作弊的用户的玩牌行为做特征提取（可根据之前实际判定为作弊的用户做样本）

### 渠道对比分析
> 对不同渠道来源的用户做整体分析，用来判断渠道的质量

* 转换为数据分析问题：
    * 对各个渠道的用户进行特征提取

### 潜在用户获取
> 对现有的高价值用户进行分析，提取出关键的特征指标，帮助渠道推广实现更加精准的投放

* 转换为数据分析问题：
    * 提取出高价值用户的固有属性和渠道特征

### 用户激活
> 挖掘未激活用户的特征

### 玩牌用户聚类
> 做牌局匹配的时候，可以让同一种类型的用户匹配在一起，例如高水平的用户匹配高水平的，中等水平匹配中等水平，从而提高用户体验

* 转换为数据分析问题：
    * 这个和精准运营是类似的问题，不同的地方在于，精准运营是离线分析，而这个更注重实时分析

### 用户路径分析
> 分析用户的游戏路径，帮助优化游戏

* 转换为数据分析问题：
    * 利用图论等分析方法，找出大部分用户进入游戏之后的行为轨迹

### 游戏体验问题
> 对用户的设备、网络等进行分析，及早发现影响用户体验的流畅度、卡顿等问题

* 转换为数据分析问题：
    * 对错误日志、用户牌局时长、是否流失等进行分析，给出体验最不佳的机型列表
